## we'll create visualizations to further explore these relationships:



    A histogram to visualize the distribution of bike prices.


```python
# Plotting the histogram of bike prices
plt.figure(figsize=(8, 5), facecolor='white')
sns.histplot(df['price'], bins=30, color='blue')
plt.title('Distribution of Bike Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

According to the histogram, most secondhand motorcycles are under $ $100000,  with a peak in frequency observed for bikes between $ $25,000 and $ $50,000. The distribution is right-skewed, meaning that some bikes have much higher pricing than the majority, which are still in the lower price range.

### creating a scatter plot to examine the relationship between bike power and price.

```python
# Plotting the scatter plot of bike power vs. price
plt.figure(figsize=(8, 5))
sns.scatterplot(x='power', y='price', data=df, color='green')
plt.title('Bike Power vs. Price')
plt.xlabel('Power (cc)')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

    Here's the scatter plot showing the relationship between bike power and price:



Bike power (measured in cc) and price show a positive trend in the scatter plot, which is consistent with the high link found in the correlation matrix. Price tends to rise in parallel with bike power, meaning that bikes with greater power are typically more expensive.

### Now visualizing the correlation matrix using a heatmap to better understand the relationships between all numerical features.

```python
# Select numerical features for the correlation matrix
numerical_data = df[['price', 'kms_driven', 'age', 'power']]

# Calculate the correlation matrix
corr_matrix = numerical_data.corr()

# Create the heatmap
plt.figure(figsize=(6, 4), facecolor='white')
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```

The associations between the dataset's numerical features are shown visually in this heatmap. The color scale makes it easy to determine the direction and strength of the associations, going from blue (positive correlation) to red (negative correlation). For instance, it is clear that price and power have a high positive link and that price and age have a negative correlation.

Price and Power (0.86): This robust positive association shows that a bike's price generally rises in parallel with its power. This association makes sense because motorcycles with greater power typically have more sophisticated features and capabilities that support their higher price.

Age and Price (-0.40): This negative connection indicates that the cost of older bikes is generally lower. This is to be expected because of depreciation, which is the process by which a bike's value declines over time as a result of deterioration, the introduction of newer models, and developments in technology.

Age and Kilometres Driven (0.52): A somewhat positive association suggests that motorcycles that are older typically have a higher mileage. This association makes sense since as bikes age, they acquire mileage.

Kms traveled and Power (0.09): A weak positive correlation indicates an approximate tendency for bikes with more power to have traveled more kilometers, but not an important connection. Numerous factors, such as the kind of use (such as city versus long-distance riding) and the owner's maintenance habits, could be to responsible for this.
